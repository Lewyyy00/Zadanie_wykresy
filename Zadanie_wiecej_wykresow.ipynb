```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]
df = pd.DataFrame(prices)
df = pd.DataFrame(prices, columns=["month", "price"])
df['priceUSD'] = df['price'].apply(lambda prices: prices / 4)
df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['month'], df['priceUSD'], 'r--')
plt.xlabel('month')
plt.ylabel('Price in USD')
plt.title('Price of goods (USD)')
plt.grid(True)
plt.show()
```
