```python
import matplotlib.pyplot as plt
```

```python
X = [0,1,2,3,4,5,6,7]
Y = [0,1,1,0,0,1,1,0]
plt.plot(X, Y)
```

```python
salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]

names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
plt.plot(salary_values)
```

```python
plt.bar(names, salary_values)
```

```python
plt.title("Salaries")
plt.xlabel("Employee")
plt.ylabel("Salary in zł")
plt.bar(names, salary_values)
```

```python
salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]

names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
plt.barh(names, salary_values, color= 'green')

plt.title("Salaries")
plt.xlabel("Salary in zł")
plt.ylabel("Names")


plt.show()
```

```python
import plotly.graph_objects as go
```

```python
X = [0,1,2,3]
Y = [0,1,0,1]

data = go.Scatter(x=X, y=Y)
layout = {
	'title': 'Zigzag with plotly'
}
fig = go.Figure(data, layout)
fig.show()
```

```python
salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]

names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))
data = go.Bar(x=names, y=salary_values)
layout = {
	'title': 'Salaries with plotly'
}
fig = go.Figure(data, layout)
fig.show()
```
